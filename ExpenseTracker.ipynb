```python
import numpy as np
import pandas as pd
from datetime import date
```

```python
SERVICES = []
PRICE = []
DATE = []
EXPENSE_TYPE = []
```

```python
def add_expense(services,price,date,expense_type):
    SERVICES.append(services)
    PRICE.append( price)
    DATE.append(date)
    EXPENSE_TYPE.append(expense_type)
    
```

```python
option = -1
while(option !=0):
    print("Welcome to expencse tracker:")
    print("1.Add food expense")
    print("2.Add houshold expense")
    print("3.Add transportation expense")
    print("4.Show and save expense")
    print("5.exit")
    option = int(input("choose option:\n"))
    
    print()
    
    if option ==5:
        print("Exiting the program")
        break
    elif option ==1:
        print("Adding food")
        expense_type = 'food'
    elif option ==2:
        print("Adding household")
        expense_type = 'household'
    elif option ==3:
        print("Adding transportation")
        expense_type = 'transportation'
    elif option ==4:
        expense_report = pd.DataFrame()
        expense_report['SERVICES'] = SERVICES
        expense_report['PRICE'] = PRICE
        expense_report['DATE'] = DATE
        expense_report['EXPENSE_TYPE'] = EXPENSE_TYPE
        
        expense_report.to_csv('expense.csv')
        
        print(expense_report)
    else:
        print("incorrect option chose.please chose correct option!")
        
    if option == 1 or option == 2 or option == 3:
        service = input("Enter the good service for the expense type"+expense_type+":\n")
        price = float(input("Enter the price:\n"))
        today = date.today()
        add_expense(service,price,today,expense_type)
        print()
        
```
